```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#曲线图
```

```python
x = np.linspace(0,14,100)
```

```python
x
```

```python
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sinplot():
    plt.plot(x, y1)
    plt.plot(x, y2)
```

```python
sinplot()
```
